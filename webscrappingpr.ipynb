```python
from bs4 import BeautifulSoup
import requests

url='https://www.scrapethissite.com/pages/forms/'

page=requests.get(url)
```

```python
Web=BeautifulSoup(page.text, 'html')
```

```python
print(Web)
```

```python
Web.find('div')
```

```python
Web.find_all('div')

#Difference between find and find_all is that find will give us the first few while find_all will give us all
```

```python
Web.find_all('div',class_="col-md-6")
```

```python
Web.find_all('p')
```

```python
Web.find_all('p',class_='lead')
```

```python
Web.find('p',class_='lead').text.strip()
```

```python
Web.find_all('th')
```

```python
Web.find('th').text.strip()
```

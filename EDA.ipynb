```python
df.groupby('Continent').mean()
```

```python
df2=df.groupby('Continent').mean().sort_values(by='World Population Percentage',ascending=False)

df2.plot.scatter(x='Rank',y='2022 Population')
```

```python
df3=df2.groupby('Continent')[['1970 Population',
       '1980 Population', '1990 Population', '2000 Population',
       '2010 Population', '2015 Population', '2020 Population',
       '2022 Population']].mean().sort_values(by="2022 Population",ascending=False)
print(df3)
```

```python
df4=df3.transpose()
df4
```

```python
df4.plot()
```

```python
df4.plot(ylabel="Rate",subplots=True)
```
